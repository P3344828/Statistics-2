```python
# Question 6: Write a Python program to generate a binomial distribution with n = 10 and p = 0.5, then plot its histogram.
# Answer:
import numpy as np
import matplotlib.pyplot as plt

# Parameters (binomial distribution)
n = 10  # no. of trials
p = 0.5 # probability of success
size = 10000 # no. of samples

# Generate binomial distribution samples
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(0, n+2)-0.5, density=True, edgecolor='white', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#  Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# Sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# Answer: At a 5% significance level, the sample mean is not significantly different from the population mean of 50. So, we fail to reject H₀.
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# Parameters
mu = 50     # Population mean
sigma = 0.5 # Population Standard deviation (known)
n = len(sample_data) # Sample size

# Sample statistics
sample_mean = np.mean(sample_data)

# Z-statistics
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"Sample mean: {sample_mean:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis")
else:
  print("Fail to reject null hypothesis")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# Answer:
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data from a normal distribution
np.random.seed(1) # for reproducibility
mu = 100 # true population mean
sigma = 15 # true population standard deviation
n = 100 # sample size
data = np.random.normal(loc=mu, scale=sigma, size=n)

# Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1) # sample standard deviation
confidence = 0.95
alpha = 1 - confidence

# t-critical value
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

# margin of error
margin_error = t_crit * (sample_std / np.sqrt(n))

# confidence interval
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='skyblue', edgecolor='blue', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')

plt.title('Histogram of Normally Distributed Sample Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# Question 9: Write a Python function to calculate the Z-Scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
# Answer: 1. Z-score of 0 means the data point is exactly at the mean
# 2. Z-scores between -1 and +1 represent values close to the mean
# 3. Z-scores < -2 or > +2 are considered unusual or outliers in a normal distribution
import numpy as np
import matplotlib.pyplot as plt

def compute_and_plot_z_scores(data):
# Convert to numpy array
  data = np.array(data)

  # Calculate mean & standard deviation
  mean = np.mean(data)
  std = np.std(data, ddof=1)

  # Compute Z-scores
  z_scores = (data - mean) / std

  # first few Z-scores
  print("First 5 Z-scores:", np.round(z_scores[:5], 3))

  # Plot histogram
  plt.figure(figsize=(10,6))
  plt.hist(z_scores, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
  plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
  plt.axvline(1, color='green', linestyle='--', label='Z = +1')
  plt.axvline(-1, color='green', linestyle='--', label='Z = -1')

  plt.title('Histogram of Z-Scores (Standardized Data)')
  plt.xlabel('Z-Score')
  plt.ylabel('Frequency')
  plt.legend()
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.tight_layout()
  plt.show()
  return z_scores

# Example
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = compute_and_plot_z_scores(sample_data)
```
